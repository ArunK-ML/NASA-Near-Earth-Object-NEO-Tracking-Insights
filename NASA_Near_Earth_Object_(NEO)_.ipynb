```python
import requests
import pprint

asteroids_data = []
target = 10000

url = "https://api.nasa.gov/neo/rest/v1/feed?start_date=2024-01-01&end_date=2024-01-07&api_key=E8laxB4HNUrbNSrlHmhopiwKuThHRsRXp1vQXGz9"
while len(asteroids_data) < target :
  response = requests.get(url)
  nasa_data = response.json()
  details = nasa_data['near_earth_objects']
  for date, asteroids in details.items():
    for ast in asteroids:
      asteroids_data.append(dict(ID = int(ast['id']),
                                 Neo_Id = ast["neo_reference_id"] ,
                                 Name = ast["name"] ,
                                 Estimated_Dia_Min = ast['estimated_diameter']['kilometers']['estimated_diameter_min'],
                                 Estimated_Dia_Max = ast['estimated_diameter']['kilometers']['estimated_diameter_max'],
                                 Close_Approach_Date = ast['close_approach_data'][0]['close_approach_date'] ,
                                 Relative_Velocity = ast['close_approach_data'][0]['relative_velocity']['kilometers_per_hour'] ,
                                 Absolute_Magnitude_h = ast["absolute_magnitude_h"] ,
                                 Astronomical = ast['close_approach_data'][0]['miss_distance']['astronomical'] ,
                                 Miss_Distance = ast['close_approach_data'][0]['miss_distance']['kilometers'] ,
                                 Miss_Distance_Lunar = ast['close_approach_data'][0]['miss_distance']['lunar'] ,
                                 Orbiting_Body = ast['close_approach_data'][0]['orbiting_body'] ,
                                 Is_Potentially_Haza = ast["is_potentially_hazardous_asteroid"]
                                 ))

      if len(asteroids_data) == target:
        break
    if len(asteroids_data) == target:
      break
  url = nasa_data['links']['next']
```

Now that we have connected to the database and selected the `NasaNeo` database, we can create a table to store the asteroid data. We will define the table schema based on the keys present in the `asteroids_data` dictionary.

```python
create_table_query = """
CREATE TABLE IF NOT EXISTS asteroids (
    ID BIGINT PRIMARY KEY,
    Neo_Id VARCHAR(255),
    Name VARCHAR(255),
    Estimated_Dia_Min DOUBLE,
    Estimated_Dia_Max DOUBLE,
    Close_Approach_Date VARCHAR(255),
    Relative_Velocity VARCHAR(255),
    Absolute_Magnitude_h DOUBLE,
    Astronomical VARCHAR(255),
    Miss_Distance VARCHAR(255),
    Miss_Distance_Lunar VARCHAR(255),
    Orbiting_Body VARCHAR(255),
    Is_Potentially_Haza BOOLEAN
)
"""

Nasa_detials.execute(create_table_query)
NASA.commit()

print("Asteroids table created successfully!")
```

Now that the `asteroids` table is created, we can insert the collected asteroid data into it.

```python
insert_query = """
INSERT INTO asteroids (
    ID, Neo_Id, Name, Estimated_Dia_Min, Estimated_Dia_Max,
    Close_Approach_Date, Relative_Velocity, Absolute_Magnitude_h,
    Astronomical, Miss_Distance, Miss_Distance_Lunar, Orbiting_Body,
    Is_Potentially_Haza
) VALUES (
    %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s
)
"""

# Prepare the data for insertion
data_to_insert = []
for asteroid in asteroids_data:
    data_to_insert.append((
        asteroid['ID'], asteroid['Neo_Id'], asteroid['Name'],
        asteroid['Estimated_Dia_Min'], asteroid['Estimated_Dia_Max'],
        asteroid['Close_Approach_Date'], asteroid['Relative_Velocity'],
        asteroid['Absolute_Magnitude_h'], asteroid['Astronomical'],
        asteroid['Miss_Distance'], asteroid['Miss_Distance_Lunar'],
        asteroid['Orbiting_Body'], asteroid['Is_Potentially_Haza']
    ))

# Execute the insert query
try:
    Nasa_detials.executemany(insert_query, data_to_insert)
    NASA.commit()
    print(f"{len(data_to_insert)} records inserted successfully!")
except Exception as e:
    NASA.rollback() # Rollback in case of error
    print(f"Error inserting data: {e}")
finally:
    Nasa_detials.close()
    NASA.close()
```
